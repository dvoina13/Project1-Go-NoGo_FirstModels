```python
def transform_intoPercept(prob_misID):
    
    p_m_real, p_b_real, trials_tot, trials_stop, trials_m, trials_b, trials_stop_m, trials_stop_b = p_real()
    percept = prob_misID*(p_b_real-p_m_real)+p_m_real  #notation: p(i_p=m) = percept !!
    
    percept_givenM = 1-prob_misID
    percept_givenB = prob_misID
    
    return percept, percept_givenM, percept_givenB
```

```python
def transform_into_prob_misID(percept):

    p_m_real, p_b_real, trials_tot, trials_stop, trials_m, trials_b, trials_stop_m, trials_stop_b = p_real()
    prob_misID = (percept-p_m_real)/(p_b_real-p_m_real)
    
    return prob_misID
```

```python
#modified function !
def probability_misidentify(x):
    
    #parameters of the model
    alpha = x[0]
    k_o = x[1]
    k_s = x[2]    
    k_d = x[3]
    x_0 = 1
    
    prob_misID_identity = np.zeros((len(orientations), len(sizes)))
    prob_misID_ori = np.zeros((len(orientations), len(sizes)))
    prob_misID_size = np.zeros((len(orientations), len(sizes)))
    
    #probabilities of misidentification, from which we infer p-stop. The prob_misID depend on s,o, s_T,o_T
    #our model for prob_msID is a hill function      
    for ind2,o in enumerate(orientations):
        for ind1,s in enumerate(sizes):
            if s!=0:
                
                #here, there are expressions not only for the probability to misidentify the identity of an image, but also 
                #probabilities to misidentify the size/orientations. Expressions are similar in form to the first.
                
                prob_misID_identity[ind2,ind1] = 0.5*1/(1+x_0/(k_s*math.fabs(s-s_T)+k_o*math.fabs(o-o_T)+k_d*(s_max-s)+alpha))
                prob_misID_ori[ind2,ind1] = 0.5*1/(1+x_0/(k_o*math.fabs(o-o_T)+alpha)) #probability to confuse the orientation
                prob_misID_size[ind2,ind1] = 0.5*1/(1+x_0/(k_s*math.fabs(s-s_T)+k_d*(s_max-s)+alpha)) #probability to confuse the size
                
                #prob_misID_identity[ind2,ind1] = k_s*math.fabs(s-s_T)+k_o*math.fabs(o-o_T)+k_d*(s_max-s)+alpha
                #prob_misID_ori[ind2,ind1] = k_o*math.fabs(o-o_T)+alpha
                #prob_misID_size[ind2,ind1] = k_s*math.fabs(s-s_T)+k_d*(s_max-s)+alpha
                
            else:
                
                prob_misID_identity[ind2,ind1] = 0.5
                prob_misID_ori[ind2,ind1] = 0.5
                prob_misID_size[ind2,ind1] = 0.5
                
    #another version of the model   
    #prob_misID[i,ind1,ind2] = 0.5*1/(1+x_0/(k_s*math.fabs(s-s_T)+k_o*math.fabs(o-o_T)+k_d*math.fabs*(s_max-s)+alpha))
    
    return prob_misID_identity, prob_misID_ori, prob_misID_size
```

```python
#modified function !
def p_stop_ProbabilityMatching(x):
    
    p_base = np.zeros((len(orientations),len(sizes)))
    p_ip_m = np.zeros((len(orientations),len(sizes)))
    p_stop_givenM_model = np.zeros((len(orientations),len(sizes)))
    p_stop_givenB_model = np.zeros((len(orientations),len(sizes)))
    
    prob_misID_identity, prob_misID_ori, prob_misID_size = probability_misidentify(x)
    p_m_real, p_b_real, trials_tot, trials_stop, trials_m, trials_b, trials_stop_m, trials_stop_b = p_real()
    
    total = len(df)
    
    for ind2,o in enumerate(orientations):
        for ind1,s in enumerate(sizes):
            
            #same thing as in the previous version for misidentification of identity
            p_ip_m[ind2,ind1] = (1-prob_misID_identity[ind2,ind1])*p_m_real[ind2,ind1]+prob_misID_identity[ind2,ind1]*p_b_real[ind2,ind1]
            
            #probability to get the orientation as a  function of the probability to misidentify
            if o>o_T:
               p_ori = 1-prob_misID_ori[ind2,ind1]
            else:
               p_ori = prob_misID_ori[ind2,ind1] 
            
            #probability to get the size as a  function of the probability to misidentify
            if s>s_T:
               p_s = 1-prob_misID_size[ind2,ind1] 
            else:
               p_s = prob_misID_size[ind2,ind1] 
                
            #total probability to consider an object rewarded given the identity, orientation o, size s    
            p_base[ind2,ind1] = p_ip_m[ind2,ind1]*(1-p_ori*p_s)
            p_stop_givenM_model[ind2,ind1] = (1-prob_misID_identity[ind2,ind1])*(1-p_ori*p_s)
            p_stop_givenB_model[ind2,ind1] = prob_misID_identity[ind2,ind1]*(1-p_ori*p_s)
            
    return p_base, p_stop_givenM_model, p_stop_givenB_model
```

```python
def p_real():
    
    p_m_real = np.zeros((len(orientations),len(sizes)))
    p_b_real = np.zeros((len(orientations),len(sizes)))
    trials_tot = np.zeros((len(orientations),len(sizes)))
    trials_stop = np.zeros((len(orientations),len(sizes)))
    trials_m = np.zeros((len(orientations),len(sizes)))
    trials_b = np.zeros((len(orientations),len(sizes)))
    trials_stop_m = np.zeros((len(orientations),len(sizes)))
    trials_stop_b = np.zeros((len(orientations),len(sizes)))
    
    for ind2,o in enumerate(orientations):
        for ind1,s in enumerate(sizes):
                
            trials_tot[ind2,ind1] = len(df[(df.Size==s)&(df.Ori==o)].total_trial_ind)
            trials_stop[ind2,ind1] = len(df[(df.Size==s)&(df.Ori==o)&(df.choice==True)].total_trial_ind)
            
            trials_m[ind2,ind1] = len(df[(df.Size==s)&(df.Ori==o)&(df.Image=='mushroom')].total_trial_ind)
            trials_b[ind2,ind1] = len(df[(df.Size==s)&(df.Ori==o)&(df.Image=='bird')].total_trial_ind)   
             
            trials_stop_m[ind2,ind1] = len(df[(df.Size==s)&(df.Ori==o)&(df.Image=='mushroom')&(df.choice==True)].total_trial_ind)
            trials_stop_b[ind2,ind1] = len(df[(df.Size==s)&(df.Ori==o)&(df.Image=='bird')&(df.choice==True)].total_trial_ind)   
            
            p_m_real[ind2, ind1] = trials_m[ind2,ind1]/trials_tot[ind2,ind1]
            p_b_real[ind2, ind1] = trials_b[ind2,ind1]/trials_tot[ind2,ind1]   
        
    return p_m_real, p_b_real, trials_tot, trials_stop, trials_m, trials_b, trials_stop_m, trials_stop_b   
```
